```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class CircularLinkedList:
    def __init__(self):
        self.head=None
    def insert_end(self,data):
        nn=Node(data)
        if self.head is None:
            self.head=nn
            nn.next=self.head
            return
        t=self.head
        while t.next!=self.head:
            t=t.next
        t.next=nn
        nn.next=self.head
    def insert_beg(self,data):
        nn=Node(data)
        if self.head is None:
            self.head=nn
            nn.next=self.head
            return
        t=self.head
        while t.next!=self.head:
            t=t.next
        nn.next=self.head
        t.next=nn
        self.head=nn
    def insert_pos(self,data,pos):
        nn=Node(data)
        if pos==1:
            self.insert_beg(data)
            return
        t=self.head
        c=1
        while t.next!=self.head and c<pos-1:
            t=t.next
            c+=1
        nn.next=t.next
        t.next=nn
    def delete_end(self):
        if self.head is None:
            print("empty")
            return
        if self.head.next==self.head:
            self.head=None
            return
        t=self.head
        while t.next.next!=self.head:
            t=t.next
        t.next=self.head
    def delete_beg(self):
        if self.head is None:
            print("Empty")
            return
        if self.head.next==self.head:
            self.head=None
            return
        t=self.head
        while t.next!=self.head:
            t=t.next
        t.next=self.head.next
        self.head=self.head.next
    def delete_value(self,value):
        if self.head is None:
            print("Empty")
            return
        t=self.head
        if t.data==value:
            self.delete_beg()
            return
        prev=None
        while t.next!=self.head and t.data!=value:
            prev=t
            t=t.next
        if t.data !=value:
            print("Not found")
            return
        prev.next=t.next
    def display(self):
        if self.head is None:
            print("Empty")
            return

        t = self.head
        while True:
            print(t.data, end=" -> ")
            t = t.next
            if t == self.head:
                break
        print("(back to head)")


# ================= MENU =================
cll = CircularLinkedList()

while True:
    print("\n--- CIRCULAR LINKED LIST MENU ---")
    print("1. Insert Beginning")
    print("2. Insert End")
    print("3. Insert Position")
    print("4. Delete Beginning")
    print("5. Delete End")
    print("6. Delete Value")
    print("7. Display")
    print("8. Exit")

    ch = int(input("Enter choice: "))

    if ch == 1:
        val = int(input("Enter value: "))
        cll.insert_beg(val)

    elif ch == 2:
        val = int(input("Enter value: "))
        cll.insert_end(val)

    elif ch == 3:
        val = int(input("Enter value: "))
        pos = int(input("Enter position: "))
        cll.insert_pos(val, pos)

    elif ch == 4:
        cll.delete_beg()

    elif ch == 5:
        cll.delete_end()

    elif ch == 6:
        val = int(input("Enter value: "))
        cll.delete_value(val)

    elif ch == 7:
        cll.display()

    elif ch == 8:
        print("Exit")
        break

    else:
        print("Invalid choice")
    
        
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None
class dlinked_list:
    def __init__(self):
        self.head=None
    def insert_end(self,data):
        nn=Node(data)
        if self.head is None:
            self.head=nn
            return
        t=self.head
        while t.next:
            t=t.next
        t.next=nn
        nn.prev=t
    def insert_beg(self,data):
        nn=Node(data)
        if self.head:
            self.head.prev=nn
            nn.next=self.head
        self.head=nn
    def insert_p(self,data,pos):
        nn=Node(data)
        if self.head is None:
            if pos == 1:
                self.head = nn
            else:
                print("Invalid position")
            return
        if pos == 1:
            self.insert_beg(data)
            return
        c=1
        t=self.head
        while t and c<pos-1:
            c+=1
            t=t.next
        if not t:
            print("Invalid")
            return
        nn.next=t.next
        nn.prev=t
        if t.next:
            t.next.prev=nn
        t.next=nn
    def delete_end(self):
        if self.head is None:
            print("Empty")
            return
        t=self.head
        if t.next is None:
            self.head=None
            return
        while t.next:
            t=t.next
        t.prev.next=None
        
    def delete_beg(self):
        if self.head is None:
            print("Empty")
            return
        t=self.head
        self.head=self.head.next
        if self.head:
            self.head.prev=None
    def delete_p(self,value):
        if self.head is None:
            print("Empty")
            return
        t=self.head
        if t and t.data==value:
            self.delete_beg()
            return
        while t and t.data!=value:
            t=t.next
        if not t:
            print("Not found")
            return
        if t.next:
            t.next.prev=t.prev
        if t.prev:
            t.prev.next=t.next
    def displayf(self):
        if self.head is None:
            print("Empty")
            return
        t=self.head
        while t:
            print(t.data,end=" ")
            t=t.next
        print("None")
    def displayb(self):
        if self.head is None:
            print("Empty")
            return
        t=self.head
        while t.next:
            t=t.next
        while t:
            print(t.data,end=" ")
            t=t.prev
        print("None")
dll = dlinked_list()

while True:
    print("\n--- DOUBLY LINKED LIST MENU ---")
    print("1. Insert Beginning")
    print("2. Insert End")
    print("3. Insert Position")
    print("4. Delete Beginning")
    print("5. Delete End")
    print("6. Delete Value")
    print("7. Display Forward")
    print("8. Display Reverse")
    print("9. Exit")

    ch = int(input("Enter choice: "))

    if ch == 1:
        val = int(input("Enter value: "))
        dll.insert_beg(val)

    elif ch == 2:
        val = int(input("Enter value: "))
        dll.insert_end(val)

    elif ch == 3:
        val = int(input("Enter value: "))
        pos = int(input("Enter position: "))
        dll.insert_p(val, pos)

    elif ch == 4:
        dll.delete_beg()

    elif ch == 5:
        dll.delete_end()

    elif ch == 6:
        val = int(input("Enter value: "))
        dll.delete_p(val)

    elif ch == 7:
        dll.displayf()

    elif ch == 8:
        dll.displayb()

    elif ch == 9:
        print("Exit")
        break

    else:
        print("Invalid choice")
        
        
        
```

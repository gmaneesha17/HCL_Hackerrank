```python
def is_safe(b,r,c,n):
    for row in range(r):
       if b[row][c]=='C':
           return False
    for col in range(c):
        if b[r][col]=='C':
            return False
    i,j=r-1,c-1
    while i>=0 and j>=0:
        if b[i][j]=='C':
            return False
        i-=1
        j-=1
    i,j=r-1,c+1
    while i>=0 and j<n:
        if b[i][j]=='C':
            return False
        i-=1
        j+=1
    i,j=r+1,c+1
    while i<n and j<n:
        if b[i][j]=='C':
            return False
        i+=1
        j+=1
    i,j=r+1,c-1
    while i<n and j>=0:
        if b[i][j]=='C':
            return False
        i+=1
        j-=1
    return True
def solve(b,r,n):
    if r == n:
        return True
    for col in range(n):
        if b[r][col]=='.' and is_safe(b,r,col,n):
            b[r][col]='C'
            if solve(b,r+1,n):
                return True
            b[r][col]='.'
    return False
n=int(input())
b=[list(input().strip()) for _ in range(n)]
if solve(b,0,n):
    for row in b:
        print(" ".join(row))
else:
    print("NO")
            
    
        
        
```

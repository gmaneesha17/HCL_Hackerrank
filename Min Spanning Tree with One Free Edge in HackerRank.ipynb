```python
def find(parent,x):
    if parent[x]!=x:
        parent[x]=find(parent,parent[x])
    return parent[x]
def union(parent,rank,x,y):
    px=find(parent,x)
    py=find(parent,y)
    if px==py:
        return False
    if rank[px]<rank[py]:
        parent[px]=py
    elif rank[px]>rank[py]:
        parent[py]=px
    else:
        parent[py]=px
        rank[px]+=1
    return True
def mst():
    n,m=map(int,input().split())
    ed=[]
    for i in range(m):
        u,v,w=map(int,input().split())
        ed.append([u,v,w])
    ans=float('inf')
    for f in range(m):
        ned=[]
        for i in range(m):
            u,v,w=ed[i]
            if f==i:
                ned.append([u,v,0])
            else:
                ned.append([u,v,w])
        ned.sort(key=lambda x :x[2])
        rank=[0]*(n+1)
        parent=[i for i in range(n+1)]
        count=0
        cost=0
        for u,v,w in ned:
            if union(parent,rank,u,v):
                cost+=w
                count+=1
        if count==n-1:
            ans=min(ans,cost)
    print(ans)
mst()
            
        
```

```python
def jumpgame(a):
    f=0
    e=0
    j=0
    for i in range(len(a)-1):
        f=max(f,i+a[i])
        if i==e:
            j+=1
            e=f
    return j
a=list(map(int,input().split()))
print(jumpgame(a))
```

```python
def rip(s):
    res=[]
    def backtrack(start,path):
        if len(path)==4:
            if start==len(s):
                res.append(".".join(path))
            return
        for i in range(1,4):
            if start+i>len(s):
                break
            part=s[start:start+i]
            if (len(part)>1 and part[0]=="0") or int(part)>255:
                continue
            path.append(part)
            backtrack(start+i,path)
            path.pop()
    backtrack(0,[])
    return res
s=input()
print(rip(s))
```

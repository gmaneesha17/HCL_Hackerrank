```python
def mrvs(streams,limit):
    streams.sort(key=lambda x:x[1]/x[0], reverse=True)
    tr=0
    u=0
    for bw,r in streams:
        if bw+u<=limit:
            u+=bw
            tr+=r
        else:
            re=limit-u
            tr+=(r/bw)*re
            break
            
    return tr
n=int(input())
s=[]
for i in range(n):
    bw,r=list(map(int,input().split()))
    s.append((bw,r))
limit=int(input())
print(mrvs(s,limit))
```

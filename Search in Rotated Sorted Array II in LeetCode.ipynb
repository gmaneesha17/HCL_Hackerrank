```python
def search(a,t):
    l=0
    h=len(a)-1
    while l<=h:
        m=(l+h)//2
        if a[m]==t:
            return True
        if a[m]==a[l]==a[h]:
            l+=1
            h-=1
            continue
        if a[l]<=a[m]:
            if a[l]<=t<=a[m]:
                h=m-1
            else:
                l=m+1
        else:
            if a[m]<t<=a[h]:
                l=m+1
            else:
                h=m-1
    return False
a=list(map(int,input().split()))
t=int(input())
print(search(a,t))
```

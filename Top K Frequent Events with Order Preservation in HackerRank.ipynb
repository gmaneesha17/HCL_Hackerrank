```python
def topkfreqe(e,k):
    freq={}
    fp={}
    for i in range(len(e)):
        x=e[i]
        if x in freq:
            freq[x]+=1
        else:
            freq[x]=1
            fp[x]=i
    a=[]
    for x in freq:
        a.append(x)
    n=len(a)
    for i in range(n):
        for j in range(0,n-i-1):
            x=a[j]
            y=a[j+1]
            if freq[x]<freq[y]:
                a[j],a[j+1]=a[j+1],a[j]
            elif freq[x]==freq[y]:
                if fp[x]>fp[y]:
                    a[j],a[j+1]=a[j+1],a[j]
    ans=[]
    for i in range(k):
        ans.append(a[i])
    return ans
e=list(map(int,input().split()))
k=int(input())
print(topkfreqe(e,k))
        
```

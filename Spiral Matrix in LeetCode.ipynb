```python
def spiral(m):
    res=[]
    l=0
    t=0
    r=len(m[0])-1
    b=len(m)-1
    while t<=b and l<=r:
        for j in range(l,r+1):
            res.append(m[t][j])
        t+=1
        for i in range(t,b+1):
            res.append(m[i][r])
        r-=1
        if t<=b:
            for j in range(r,l-1,-1):
                res.append(m[b][j])
            b-=1
        if l<=r:
            for i in range(b,t-1,-1):
                res.append(m[i][l])
            l+=1
    return res
r = int(input("Enter rows: "))
c = int(input("Enter cols: "))

matrix = []

for i in range(r):
    row = list(map(int, input().split()))
    matrix.append(row)

print(spiral(matrix))
            
            
```

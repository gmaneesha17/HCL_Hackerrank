```python
def lcp(st):
    if not st:
        return ""
    p=""
    for i in range(len(st[0])):
        ch=st[0][i]
        for s in st[1:]:
            if i==len(s) or s[i]!=ch:
                return p
        p+=ch
    return p
n=int(input())
s=[]
for i in range(n):
    s.append(input())
print(lcp(s))
```

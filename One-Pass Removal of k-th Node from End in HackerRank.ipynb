```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
def removekthnode(head,k):
    dummy=Node(0)
    dummy.next=head
    fast=dummy
    slow=dummy
    for _ in range(k+1):
        fast=fast.next
    while fast:
        fast=fast.next
        slow=slow.next
    slow.next=slow.next.next
    return dummy.next
def plist(head):
    t=head
    while t:
        print(t.data)
        t=t.next
a=list(map(int,input().split()))
h=Node(a[0])
temp=h
for n in a[1:]:
    temp.next=Node(n)
    temp=temp.next
k=int(input())
h=removekthnode(h,k)
print(plist(h))
```

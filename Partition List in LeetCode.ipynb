```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
def partition(head,x):
    before=Node(0)
    after=Node(0)
    b=before
    a=after
    while head:
        if head.data<x:
            b.next=head
            b=b.next
        else:
            a.next=head
            a=a.next
        head=head.next
    a.next=None
    b.next=after.next
    return before.next
a=list(map(int,input().split()))
x=int(input())
head=Node(a[0])
t=head
for i in range(1,len(a)):
    t.next=Node(a[i])
    t=t.next
head=partition(head,x)
cur=head
while cur:
    print(cur.data,end=" ")
    cur=cur.next
```

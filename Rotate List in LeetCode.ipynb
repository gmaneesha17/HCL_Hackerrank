```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
def rotate(head,k):
    if not head or not head.next or k==0:
        return head
    tail=head
    n=1
    while tail.next:
        tail=tail.next
        n+=1
    tail.next=head
    k=k%n
    steps=n-k-1
    nt=head
    for i in range(steps):
        nt=nt.next
    nh=nt.next
    nt.next=None
    return nh
a=list(map(int,input().split()))
n=len(a)
head=Node(a[0])
t=head
for i in range(1,n):
    t.next=Node(a[i])
    t=t.next
k=int(input())
head=rotate(head,k)
c=head
while c:
    print(c.data,end=" ")
    c=c.next
    
```

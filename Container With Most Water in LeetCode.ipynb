```python
def cwm(height):
    l=0
    h=len(height)-1
    ma=0
    while l<h:
        w=h-1
        ht=min(height[l],height[h])
        a=w*ht
        ma=max(ma,a)
        if height[l]<height[h]:
            l+=1
        else:
            h-=1
    return ma
height=list(map(int,input().split()))
print(cwm(height))
        
```

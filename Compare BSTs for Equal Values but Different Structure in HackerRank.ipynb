```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def insert(root,data):
    if root is None:
        return Node(data)
    if data<root.data:
        root.left=insert(root.left,data)
    else:
        root.right=insert(root.right,data)
    return root
def inorder(root,a):
    if root:
        inorder(root.left,a)
        a.append(root.data)
        inorder(root.right,a)
def isequal(root1,root2):
    a1=[]
    a2=[]
    inorder(root1,a1)
    inorder(root2,a2)
    return a1==a2
n=int(input())
a=list(map(int,input().split()))
root1=None
for i in a:
    root1=insert(root1,i)
m=int(input())
b=list(map(int,input().split()))
root2=None
for l in b:
    root2=insert(root2,l)
print(isequal(root1,root2))
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def insert(root,data):
    if root is None:
        return Node(data)
    if data<root.data:
        root.left=insert(root.left,data)
    else:
        root.right=insert(root.right,data)
    return root
def issame(p,q):
    if p is None and q is None:
        return True
    if p is None or q is None or p.data!=q.data:
        return False
    return issame(p.left,q.left) and issame(p.right,q.right)
n=int(input())
a=list(map(int,input().split()))
root1=None
for i in a:
    root1=insert(root1,i)
m=int(input())
b=list(map(int,input().split()))
root2=None
for i in b:
    root2=insert(root2,i)
print(issame(root1,root2))
```

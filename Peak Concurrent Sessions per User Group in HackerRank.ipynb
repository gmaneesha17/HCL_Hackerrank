```python
def peaks():
    n=int(input())
    sessions=[]
    for i in range(n):
        g=input("Enter grp:")
        st=int(input("Enter start time"))
        end=int(input("Enter end time"))
        sessions.append([g,st,end])
    grps={}
    for s in sessions:
        g=s[0]
        st=s[1]
        end=s[2]
        if g not in grps:
            grps[g]=[]
        grps[g].append([st,1])
        grps[g].append([end,-1])
    r={}
    for g in grps:
        e=grps[g]
        n=len(e)
        for i in range(n):
            for j in range(n-i-1):
                if e[j][0]>e[j+1][0]:
                    e[j],e[j+1]=e[j+1],e[j]
        cur=0
        peak=0
        for ev in e:
            cur+=ev[1]
            if cur>peak:
                peak=cur
        r[g]=peak
    for i in r:
        print(i,":",r[i])
peaks()
            
```

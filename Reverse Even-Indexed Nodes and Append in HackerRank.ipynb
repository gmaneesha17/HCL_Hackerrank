```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
def reverse(head):
    cur=head
    prev=None
    while cur:
        nxt=cur.next
        cur.next=prev
        prev=cur
        cur=nxt
    return prev
def solve(head):
    cur=head
    eh=et=None
    oh=ot=None
    if not head or not head.next:
        return head
    index=0
    while cur:
        nxt=cur.next
        cur.next=None
        if index%2==0:
            if not eh:
                eh=et=cur
            else:
                et.next=cur
                et=cur
        else:
            if not oh:
                oh=ot=cur
            else:
                ot.next=cur
                ot=cur
        cur=nxt
        index+=1
    eh=reverse(eh)
    ot.next=eh
    return oh
n=int(input())
v=list(map(int,input().split()))
head=Node(v[0])
t=head
for i in range(1,n):
    t.next=Node(v[i])
    t=t.next
r=solve(head)
while r:
    print(r.data,end=" ")
    r=r.next
    
```

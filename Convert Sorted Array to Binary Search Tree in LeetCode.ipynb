```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def soartobst(a):
    if not a:
        return None
    m=len(a)//2
    root=Node(a[m])
    root.left=soartobst(a[:m])
    root.right=soartobst(a[m+1:])
    return root
def inorder(root):
    if root:
        inorder(root.left)
        print(root.data,end=" ")
        inorder(root.right)
n=int(input())
a=list(map(int,input().split()))
root=soartobst(a)
inorder(root)
```

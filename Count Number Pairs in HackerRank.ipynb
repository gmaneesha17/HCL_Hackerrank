```python
#diff of 2 no ==k    a-b=k   or     a=k+b   search a in arr or not 
def count(a,k):
    c=0
    for i in range(len(a)):
        for j in range(i+1,len(a)):
            dif=a[i]-a[j]
            if dif<0:
                dif=-dif
            if dif==k:
                c+=1
                print((a[i],a[j]))
    return c
a=list(map(int,input().split()))
k=int(input())
print(count(a,k))
    
```

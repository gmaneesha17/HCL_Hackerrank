```python
def zstsw(a,k):
    r=[]
    n=len(a)
    for i in range(n-k+1):
        window=a[i:i+k]
        c=0
        for j in range(k-2):
            l=j+1
            h=k-1
            while(l<h):
                t=window[j]+window[l]+window[h]
                if t==0:
                    c+=1
                    l+=1
                    h-=1
                elif t<0:
                    l+=1
                else:
                    h-=1
        r.append(c)
    return r
a=list(map(int,input().split()))
k=int(input())
print(zstsw(a,k))
                    
```

```python
def psearch(a,t):
    l=0
    h=len(a)-1
    while l<=h:
        m=(l+h)//2
        if a[m]==t:
            return m
        if a[l]<=a[m]:
            if a[l]<=t<a[m]:
                h=m-1
            else:
                l=m+1
        else:
            if a[m]<t<=a[h]:
                l=m+1
            else:
                h=m-1
    return -1
a=list(map(int,input().split()))
t=int(input())
print(psearch(a,t))
```

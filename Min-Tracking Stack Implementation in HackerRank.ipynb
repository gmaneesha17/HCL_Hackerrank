```python
class Stack:
    def __init__(self):
        self.st=[]
        self.mst=[]
    def push(self,x):
        self.st.append(x)
        if not self.mst or self.mst[-1]>=x:
            self.mst.append(x)
    def pop(self):
        if self.st:
            if self.st[-1]==self.mst[-1]:
                self.mst.pop()
            return self.st.pop()
    def minel(self):
        if self.mst:
            return self.mst[-1]
    def peek(self):
        if self.st:
            return self.st[-1]
s=Stack()
q=int(input())
for i in range(q):
    op=input().split()
    if op[0]=="Push":
        s.push(int(op[1]))
    elif op[0] == "pop":
        s.pop()
    elif op[0] == "top":
        print(s.peek())

    elif op[0] == "min":
        print(s.minel())
            
```

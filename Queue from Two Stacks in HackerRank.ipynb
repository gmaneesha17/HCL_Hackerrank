```python
def qfts():
    q=int(input())
    s1=[]
    s2=[]
    for i in range(q):
        op=input().split()
        if op[0]=='1':
            s1.append(int(op[1]))
        elif op[0]=='2':
            if not s2:
                while s1:
                    s2.append(s1.pop())
            s2.pop()
        elif op[0]=='3':
            if not s2:
                while s1:
                    s2.append(s1.pop())
            print(s2[-1])
qfts()
```

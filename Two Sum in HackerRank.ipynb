```python
def twosum(a,t):
    for i in range(len(a)):
        for j in range(i+1,len(a)):
            if a[j]+a[i]==t:
                return [i,j]
a=list(map(int,input().split()))
t=int(input())
print(twosum(a,t))
```

```python
def two_sum(a,t):
    seen={}
    for i in range(len(a)):
        n=a[i]
        needed=t-n
        if needed in seen:
            return [seen[needed],i]
        seen[n]=i
a=list(map(int,input().split()))
t=int(input())
print(two_sum(a,t))
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def recoverTree(root):
    first = [None]
    second = [None]
    prev = [TreeNode(float('-inf'))]

    def inorder(node):
        if not node:
            return
        
        inorder(node.left)

        if prev[0].val > node.val:
            if not first[0]:
                first[0] = prev[0]
            second[0] = node

        prev[0] = node
        inorder(node.right)

    inorder(root)
    first[0].val, second[0].val = second[0].val, first[0].val


def inorder_print(root):
    if not root:
        return
    inorder_print(root.left)
    print(root.val, end=" ")
    inorder_print(root.right)
print("Using fixed BST example...")
root = TreeNode(3)
root.left = TreeNode(1)
root.right = TreeNode(4)
root.right.left = TreeNode(2)

recoverTree(root)

print("Inorder after recovery:")
inorder_print(root)
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def create(root,data):
    if root is None:
        return Node(data)
    if data<root.data:
        root.left=create(root.left,data)
    else:
        root.right=create(root.right,data)
    return root
def maxdepth(root):
    if root is None:
        return 0
    l=maxdepth(root.left)
    r=maxdepth(root.right)
    return max(l,r)+1
n=int(input())
a=list(map(int,input().split()))
root=None
for n in a:
    root=create(root,n)
print(maxdepth(root))
```

```python
# depth from root to target node
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def create(root,data):
    if root is None:
        return Node(data)
    if data<root.data:
        root.left=create(root.left,data)
    else:
        root.right=create(root.right,data)
    return root
def depth(root,t,d=0):
    if root is None:
        return -1
    if root.data==t:
        return d
    elif root.data>t:
        return depth(root.left,t,d+1)
    else:
        return depth(root.right,t,d+1)
n=int(input())
a=list(map(int,input().split()))
root=None
for n in a:
    root=create(root,n)
t=int(input())
print(depth(root,t))
```

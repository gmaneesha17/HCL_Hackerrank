```python
def multiply(s1,s2):
    n=len(s1)
    m=len(s2)
    res=[0]*(m+n)
    if s1=="0" or s2=="0":
        return "0"
    for i in range(n-1,-1,-1):
        for j in range(m-1,-1,-1):
            mu=int(s1[i])*int(s2[j])
            p1=i+j
            p2=i+j+1
            t=mu+res[p2]
            res[p2]=t%10
            res[p1]+=t//10
    ans=""
    for x in res:
        if not (ans=="" and x==0):
            ans+=str(x)
    return ans
s1=input()
s2=input()
print(multiply(s1,s2))
```

```python
import sys
sys.setrecursionlimit(10**6)
def iscycle(node,g,vis,rst):
    vis[node]=True
    rst[node]=True
    for nei in g[node]:
        if not vis[nei]:
            if iscycle(nei,g,vis,rst):
                return True
        elif rst[nei]:
            return True
    rst[node]=False
    return False
def detect(n,e):
    g={i: [] for i in range(n)}
    for u,v in e:
        g[u].append(v)
    vis=[False]*n
    rst=[False]*n
    for i in range(n):
        if not vis[i]:
            if iscycle(i,g,vis,rst):
                return True
    return False
n=int(input())
m=int(input())
e=[]
for i in range(m):
    u,v=list(map(int,input().split()))
    e.append((u,v))
if detect(n,e):
    print("Cycle detected")
else:
    print("No cycle")
```

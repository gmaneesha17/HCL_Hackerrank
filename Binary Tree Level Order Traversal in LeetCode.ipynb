```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def create(root,data):
    if root is None:
        return Node(data)
    if data<root.data:
        root.left=create(root.left,data)
    else:
        root.right=create(root.right,data)
    return root
def height(root):
    if root is None:
        return 0
    l=height(root.left)
    r=height(root.right)
    return max(l,r)+1
def level(root,l):
    if root is None:
        return
    if l==1:
        print(root.data,end=" ")
    else:
        level(root.left,l-1)
        level(root.right,l-1)
def lord(root):
    h=height(root)
    for i in range(1,h+1):
        level(root,i)
        print()
n=int(input())
a=list(map(int,input().split()))
root=None
for n in a:
    root=create(root,n)
lord(root)
```

```python
def coinchange(plans,t):
    dp=[float('inf')]*(t+1)
    dp[0]=0
    for p in plans:
        for i in range(p,t+1):
            dp[i]=min(dp[i],dp[i-p]+1)
    return -1 if dp[t]==float('inf') else dp[t]
plans=list(map(int,input().split()))
t=int(input())
print(coinchange(plans,t))
        
```

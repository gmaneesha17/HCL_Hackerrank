```python
def la(s,k):
    def solve(st):
        l=0
        ml=0
        f=0
        for r in range(len(s)):
            e=str(st^(r%2))
            if s[r]!=e:
                f+=1
            while f>k:
                el=str(st^(l%2))
                if s[l]!=el:
                    f-=1
                l+=1
            ml=max(ml,r-l+1)
        return ml
    return max(solve(0),solve(1))
s = input().strip()
k = int(input())

print(la(s, k))
```

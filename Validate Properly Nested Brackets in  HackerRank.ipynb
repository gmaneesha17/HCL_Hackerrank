```python
def isbalanced(s):

    st=[]
    for ch in s:
        if ch in "{[(":
            st.append(ch)
        else:
            if not st:
                return "NO"
            top=st.pop()
            if ch==')' and top!='(':
                return "NO"
            if ch=='}' and top !='{':
                return "NO"
            if ch==']' and top!='[':
                return "NO"
    return "YES" if not st else "NO"
s=input()
print(isbalanced(s))
```

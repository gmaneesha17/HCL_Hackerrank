```python
import heapq
def shortpath():
    n,m=map(int,input().split())
    g={}
    delay=list(map(int,input().split()))
    for i in range(1,n+1):
        g[i]=[]
    for j in range(m):
        u,v,w=map(int,input().split())
        g[u].append([v,w])
        g[v].append([u,w])
    s,d=map(int,input().split())
    dist=[float('inf')]*(n+1)
    dist[s]=delay[s-1]
    pq=[]
    heapq.heappush(pq,[dist[s],s])
    while pq:
        dis,node=heapq.heappop(pq)
        if dis>dist[node]:
            continue
        for nei,w in g[node]:
            newcost=dis+w+delay[nei-1]
            if newcost<dist[nei]:
                dist[nei]=newcost
                heapq.heappush(pq,[newcost,nei])
    if dist[d]==float('inf'):
        print(-1)
    else:
        print(dist[d])
shortpath()
    
```

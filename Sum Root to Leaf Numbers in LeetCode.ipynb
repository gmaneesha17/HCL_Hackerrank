```python
from collections import deque

class TreeNode:
    def __init__(self, val=0):
        self.val = val
        self.left = None
        self.right = None

def build_tree(data):
    if not data or data[0] == "null":
        return None
    
    root = TreeNode(int(data[0]))
    q = deque([root])
    i = 1
    
    while q and i < len(data):
        node = q.popleft()

        if i < len(data) and data[i] != "null":
            node.left = TreeNode(int(data[i]))
            q.append(node.left)
        i += 1

        if i < len(data) and data[i] != "null":
            node.right = TreeNode(int(data[i]))
            q.append(node.right)
        i += 1

    return root

def sumNumbers(root):
    def dfs(node, cur):
        if not node:
            return 0
        
        cur = cur * 10 + node.val
        
        if not node.left and not node.right:
            return cur
        
        return dfs(node.left, cur) + dfs(node.right, cur)
    
    return dfs(root, 0)

data = input("Enter tree (level order with space, use null): ").split()
root = build_tree(data)
print("Sum of root-to-leaf numbers:", sumNumbers(root))
```

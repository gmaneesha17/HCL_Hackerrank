```python
#  left subtree is mirror image of right sub tree
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def insert(arr, i=0):

    if i >= len(arr) or arr[i] is None:
        return None

    root = Node(arr[i])

    root.left = insert(arr, 2*i + 1)
    root.right = insert(arr, 2*i + 2)

    return root
def ismirror(left,right):
    if left is None and right is None:
        return True
    if left is None or right is None or left.data!=right.data:
        return False
    return (ismirror(left.left,right.right) and ismirror(left.right,right.left))
def issym(root):
    if root is None:
        return True
    return ismirror(root.left,root.right)
arr = input().split()

for i in range(len(arr)):
    if arr[i] == "null":
        arr[i] = None
    else:
        arr[i] = int(arr[i])

root = insert(arr)

print(issym(root))
```

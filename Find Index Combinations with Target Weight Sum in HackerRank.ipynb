```python
def combinations(w,t):
    d={}
    for i in range(len(w)):
        need=t-w[i]
        if need in d:
            return [d[need],i]
        d[w[i]]=i
    return []
w=list(map(int,input().split()))
t=int(input())
print(combinations(w,t))
```

```python
def combinations(w,t):
    for i in range(len(w)):
        for j in range(i+1,len(w)):
            if w[i]+w[j]==t:
                return [i,j]
w=list(map(int,input().split()))
t=int(input())
print(combinations(w,t))
```

```python
def dfs(node,adj,vis):
    vis[node]=True
    for nei in adj[node]:
        if not vis[nei]:
            dfs(nei,adj,vis)
def count(n,e):
    adj=[[] for _ in range(n)]
    for u,v in e:
        adj[u].append(v)
        adj[v].append(u)
    vis=[False]*n
    c=0
    for i in range(n):
        if not vis[i]:
            dfs(i,adj,vis)
            c+=1
    return c
n=int(input())
e=int(input())
ed=[]
for i in range(e):
    u,v=list(map(int,input().split()))
    ed.append((u,v))
print(count(n,ed))
    
```

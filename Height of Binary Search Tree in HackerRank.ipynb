```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def create(root,data):
    if root is None:
        return Node(data)
    if data<root.data:
        root.left=create(root.left,data)
    else:
        root.right=create(root.right,data)
    return root
def height(root):
    if root is None:
        return -1
    l=height(root.left)
    r=height(root.right)
    return max(l,r)+1
n=int(input())
a=list(map(int,input().split()))
root=None
for n in a:
    root=create(root,n)
print(height(root))
```

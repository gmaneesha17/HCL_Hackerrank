```python
def merged_sort(intervals):
    intervals.sort()
    merged=[intervals[0]]
    for c in intervals[1:]:
        last=merged[-1]
        if c[0]<=last[1]:
            last[1]=max(c[1],last[1])
        else:
            merged.append(c)
    return merged
n=int(input())
intervals=[]
for i in range(n):
    s,e=list(map(int,input().split()))
    intervals.append([s,e])
r=merged_sort(intervals)
for i in r:
    print(i)
```
